## Korelasyon Grafikleri

#### Scatterplot

##### **total_bill :** yemeğin toplam fiyatı (bahşiş ve vergi dahil)
##### **tip :** bahşiş
##### **sex :** ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
##### **smoker :** grupta sigara içen var mı? (0=No, 1=Yes)
##### **day :** gün (3=Thur, 4=Fri, =Sat, 6=Sun)
##### **time :** ne zaman? (0=Day, 1=Night)
##### **size :** grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
# Değişken kategorik ise barplot(sütun grafik) kullandık.
# Değişken sayısal ise histogram (distplot), yoğunluk (kdeplot), boxplot ve violin grafikleri kullandık.
# Bunlara eklemiş olduğumuz 2-3. değişkenler çaprazlama anlamında yapılan eklemelerdi.
# Bu bölümde 2 yapısal değişkenin birlikte oluşturdukları yapıları inceleyeceğiz.
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip",hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", 
                y = "tip",hue = "time", 
                style = "time",
                data = df);
```

```python
sns.scatterplot(x = "total_bill", 
                y = "tip",
                hue = "day", 
                style = "time",
                data = df);
```

```python
sns.scatterplot(x = "total_bill", 
                y = "tip",
                size = "size",
               
                data = df);
```

```python
sns.scatterplot(x = "total_bill", 
                y = "tip",
                size = "size",
                hue = "size",
                data = df);
```
