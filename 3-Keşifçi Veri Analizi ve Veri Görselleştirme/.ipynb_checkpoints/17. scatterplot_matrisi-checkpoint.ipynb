## Scatterplot Matrisi

```python
import seaborn as sns

iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df);
```

```python
# Eğer oluşturulmuş olan grafik toz bulutu şeklinde ise veya
# yapısal bir formu yoksa, bu o iki değişken arasında ilişki
# olmadığını gösterir
```

```python
sns.pairplot(df, hue = "species");
```

```python
sns.pairplot(df, hue = "species", markers = ["o","s","D"]);
```

```python
sns.pairplot(df, kind = "reg");
```

```python
sns.pairplot(df, kind = "reg", hue = "species");
```
