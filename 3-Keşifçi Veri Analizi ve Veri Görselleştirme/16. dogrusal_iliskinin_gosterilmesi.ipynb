## Doğrusal İlişkinin Gösterilmesi

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df);
```

```python
sns.lmplot(x = "total_bill", 
           y = "tip", 
           hue = "smoker", 
           data = df);
```

```python
sns.lmplot(x = "total_bill", 
           y = "tip", 
           hue = "smoker",
           col = "time",
           data = df);
```

```python
sns.lmplot(x = "total_bill", 
           y = "tip", 
           hue = "smoker",
           col = "time",
           row = "sex",
           data = df);
```
