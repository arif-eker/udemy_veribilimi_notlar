## Scatterplot Matrisi

```python
import seaborn as sns

iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df);
```

```python
# Eer oluturulmu olan grafik toz bulutu eklinde ise veya
# yapsal bir formu yoksa, bu o iki deiken arasnda iliki
# olmadn gsterir
```

```python
sns.pairplot(df, hue = "species");
```

```python
sns.pairplot(df, hue = "species", markers = ["o","s","D"]);
```

```python
sns.pairplot(df, kind = "reg");
```

```python
sns.pairplot(df, kind = "reg", hue = "species");
```
